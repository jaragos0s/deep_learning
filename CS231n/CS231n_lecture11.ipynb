# Lecture 11  Detection and Segmentation

## Sementic Segmentation

Does not differentiate instances, only care about pixelf

Label each pixel in the image with a category label

Network as a bunch of convolutional layers, with downsampling and upsampling inside the network

- Downsampling : Pooling, strided convolution
- Upsampling : Unpooling or strided transpose convolution 

## Object Detection + Deep Learning

### Sliding Window

Apply a CNN to many different crops of the image, CNN classifies each crop as object or background

### Region Perposals

Find blobby image regions that are likely to contain objects


### R-CNN

Warped image regions -> Forward each region through ConvNet -> Classify regions with SVMs

### Fast R-CNN

Rather than processing CNN each ROI seperately, run the entire image through CNN -> high resolution convolution feature map

### Faster R-CNN

Make CNN do proposals

Insert Region Proposal Network to predict proposals from features

### Mask R-CNN

Faster R-CNN + mask branch + FPN, RoI align
