# Lab 05 Logistic Regression

Hypothesis

$$ H(X) = 1 / (1 + e^{-W^TX}) $$

Cost
$$ cost(W) = -1/m\sum ylog(H(x)) + (1-y)(log(1-H(x)) $$

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
# For reproducibility
torch.manual_seed(1)
```

```python
x_data = [[1, 2], [2, 3], [3, 1], [4, 3], [5, 3], [6, 2]]
y_data = [[0],[0],[0],[1],[1],[1] ]
```

```python
x_train = torch.FloatTensor(x_data)
y_train = torch.FloatTensor(y_data)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
print('e^1 equals : ', torch.exp(torch.FloatTensor([1])))
```

```python
W = torch.zeros((2, 1), requires_grad = True)
b = torch.zeros(1, requires_grad = True)
```

```python
hypothesis = 1 / (1 + torch.exp(-(x_train.matmul(W) + b)))
```

```python
print(hypothesis)
print(hypothesis.shape)
```

```python
print('1/(1+e^{-1}) equals: ', torch.sigmoid(torch.FloatTensor([1])))
```

```python
hypothesis = torch.sigmoid(x_train.matmul(W) + b)
```

```python
print(hypothesis)
print(hypothesis.shape)
```

```python
# computing the cost function 
# For one element
-(y_train[0] * torch.log(hypothesis[0]) + (1 - y_train[0]) * torch.log(1 - hypothesis[0]))
```

```python
losses = -(y_train * torch.log(hypothesis) + (1 - y_train) * torch.log(1 - hypothesis))
print(losses)
```

```python
cost = losses.mean()
print(cost)
```

```python
F.binary_cross_entropy(hypothesis, y_train)
```

```python
# initialize model
W = torch.zeros((2, 1), requires_grad= True)
b = torch.zeros(1, requires_grad= True)

# set optimizer
optimizer = optim.SGD([W, b], lr = 1)

nb_epochs = 1000
for epoch in range(nb_epochs + 1):

  # compute cost
  hypothesis = torch.sigmoid(x_train.matmul(W) + b)
  cost = F.binary_cross_entropy(hypothesis, y_train)

  # upgrade H(x) using cost
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  # print log
  if epoch % 100 == 0:
    print('Epoch {:4d}/{} Cost: {:.6f}'.format(epoch, nb_epochs, cost.item()))
```

```python
class BinaryClassifier(nn.Module):
  def __init__(self):
    super().__init__()
    self.linear = nn.Linear(8, 1) # self.linear = {W, b}
    self.sigmoid = nn.Sigmoid()

  def forward(self, x):
    return self.sigmoid(self.linear(x))
```

```python
model = BinaryClassifier()
```

```python
x_train = torch.FloatTensor([[-0.2941,  0.4874,  0.1803, -0.2929,  0.0000,  0.0015, -0.5312, -0.0333],
        [-0.8824, -0.1457,  0.0820, -0.4141,  0.0000, -0.2072, -0.7669, -0.6667],
        [-0.0588,  0.8392,  0.0492,  0.0000,  0.0000, -0.3055, -0.4927, -0.6333],
        [-0.8824, -0.1055,  0.0820, -0.5354, -0.7778, -0.1624, -0.9240,  0.0000],
        [ 0.0000,  0.3769, -0.3443, -0.2929, -0.6028,  0.2846,  0.8873, -0.6000]])
y_train = torch.FloatTensor([[0.],
        [1.],
        [0.],
        [1.],
        [0.]])
```

```python
optimizer = optim.SGD(model.parameters(), lr = 1)

nb_epochs = 100
for epoch in range(nb_epochs + 1):

  # compute H(x)
  hypothesis = model(x_train)

  # compute cost
  cost = F.binary_cross_entropy(hypothesis, y_train)

  # comput H(x) using cost
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  # print log
  if epoch % 10 == 0:
    prediction = hypothesis >= torch.FloatTensor([0.5])
    correct_prediction = prediction.float() == y_train
    accuracy = correct_prediction.sum().item() / len(correct_prediction)
    print('Epoch {:4d}/{} Cost : {:.6f} Accuracy {:2.2f}%'.format(
        epoch, nb_epochs, cost.item(), accuracy * 100
    ))
```
