## Softmax
$$P(class = i) = e^i / \sum e^i $$

```python
z = torch.FloatTensor([1,2,3])
```

```python
hypothesis = F.softmax(z, dim = 0)
print(hypothesis)
```

Since they are probabilities, they should add up to 1.

```python
hypothesis.sum()
```

## Cross Entropy

$$H(P, Q) = - \mathbb{E}_{X \sim P(X)} \left[ log Q(x) \right] =  - \sum_{x \in \mathcal(X)} P(x) logQ(x) $$


**The cross entropy loss**
$$ L = 1/N \sum -y log(\hat{y}) $$

where $\hat{y}$ is the predicted probability and y is the correct probability (0 or 1).

```python
z = torch.rand(3, 5, requires_grad= True)
hypothesis = F.softmax(z, dim = 1)
print(hypothesis)
```

```python
y = torch.randint(5, (3, )).long()
print(y)
```

```python
y_one_hot = torch.zeros_like(hypothesis)
y_one_hot.scatter_(1, y.unsqueeze(1), 1) # inplace #|y| = (3,) -> (3, 1)
```

```python
cost = (y_one_hot * -torch.log(hypothesis)).sum(dim = 1).mean()
print(cost)
```

## Cross-entropy Loss with torch.nn.functional

```python
# Low level
torch.log(F.softmax(z, dim = 1))
```

```python
# High level
F.log_softmax(z, dim = 1)
```

```python
# Low level
(y_one_hot * -torch.log(F.softmax(z, dim = 1))).sum(dim = 1).mean()
```

```python
# High level
F.nll_loss(F.log_softmax(z, dim = 1), y) # negative log likelihood
```

F.cross_entropy <- F.log_softmax() and F.nll_loss()

```python
F.cross_entropy(z, y)
```

```python
x_train = [[1, 2, 1, 1],
         [2, 1, 3, 2],
         [3, 1, 3, 4],
         [4, 1, 5, 5],
         [1, 7, 5, 5],
         [1, 2, 5, 6],
         [1, 6, 6, 6],
         [1, 7, 7, 7]]
y_train = [2, 2, 2, 1, 1, 1, 0, 0]
x_train = torch.FloatTensor(x_train)
y_train = torch.LongTensor(y_train)
```

Training

```python
# initialize model
W = torch.zeros((4, 3), requires_grad = True)
b = torch.zeros(1, requires_grad= True)
# set optimizer
optimizer = optim.SGD([W, b], lr = 0.1)

nb_epochs = 1000
for epoch in range(nb_epochs + 1):

  # compute cost
  hypothesis = F.softmax(x_train.matmul(W) + b, dim = 1) 
  y_one_hot = torch.zeros_like(hypothesis)
  y_one_hot.scatter_(1, y_train.unsqueeze(1), 1)
  cost = (y_one_hot * -torch.log(F.softmax(hypothesis, dim = 1))).sum(dim = 1).mean()

  # update H(x) using cost
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  # print log per 100
  if epoch % 100 == 0:
    print('Epoch {:4d} / {} Cost: {:.6f}'.format(
        epoch, nb_epochs, cost.item()
    ))
```

```python
# initialize model
W = torch.zeros((4, 3), requires_grad = True)
b = torch.zeros(1, requires_grad= True)
# set optimizer
optimizer = optim.SGD([W, b], lr = 0.1)

nb_epochs = 1000
for epoch in range(nb_epochs + 1):

  # compute cost
  z = x_train.matmul(W) + b
  cost = F.cross_entropy(z, y_train)

  # update H(x) using cost
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  # print log per 100
  if epoch % 100 == 0:
    print('Epoch {:4d} / {} Cost: {:.6f}'.format(
        epoch, nb_epochs, cost.item()
    ))
```

High-level implementation with nn.Module

```python
class SoftmaxClassifierModel(nn.Module):
  def __init__(self):
    super().__init__()
    self.linear = nn.Linear(4, 3) # output is 3

  def forward(self, x):
    return self.linear(x)
```

```python
model = SoftmaxClassifierModel()
```

```python
optimizer = optim.SGD(model.parameters(), lr = 0.1)

nb_epochs = 1000
for epoch in range(nb_epochs + 1):

  # compute cost
  prediction = model(x_train)
  cost = F.cross_entropy(prediction, y_train)

  # update H(x) using cost
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  # print log per 100
  if epoch % 100 == 0:
    print('Epoch {:4d} / {} Cost: {:.6f}'.format(
        epoch, nb_epochs, cost.item()
    ))
```
